```python
filtered_nextstrain = nextstrain[(nextstrain['genbank_accession'] != '?') & (nextstrain['date'] != '?')]
```

```python
filtered_nextstrain
```

```python
filtered_nextstrain['Nextstrain_clade'].value_counts()
```

```python
merged_data = cov_metadata.merge(filtered_nextstrain, left_on="Accession", right_on="genbank_accession", how="left")
```

```python
merged_data
```

```python
merged_data['Nextstrain_clade'].value_counts()
```

```python
first_last = pd.concat([
    merged_data.loc[merged_data.groupby('Nextstrain_clade')['Release_Date'].idxmin()],
    merged_data.loc[merged_data.groupby('Nextstrain_clade')['Release_Date'].idxmax()]
]).sort_values(['Nextstrain_clade', 'Release_Date'])
```
